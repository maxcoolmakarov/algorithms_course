### Пример 1

1. Подготовка и подача искового заявления (для заявителя)

/new_claim: начало создания нового искового заявления. 

/upload_document: загрузка искового заявления

/add_attachment: добавление дополнительных документов к исковому заявлению

/check_claim: проверка искового заявления на соответствие формальным требованиям

/preview_claim: предварительный просмотр искового заявления перед отправкой

/submit_claim: отправка искового заявления в суд

/claim_status [номер дела]: проверка статуса искового заявления по номеру дела.


2. Прием и обработка искового заявления (для сотрудника суда)

/receive_claim: прием искового заявления

/assign_claim [номер дела] [судья/служащий]: назначение дела конкретному судье или сотруднику суда

/register_claim [номер дела]: регистрация искового заявления в системе 

/review_claim [номер дела]: просмотр и проверка искового заявления

/reject_claim [номер дела] [причина]: отклонение искового заявления с указанием причины
или
/accept_claim [номер дела]: принятие искового заявления

/forward_claim [номер дела] [подразделение]: передача искового заявления в другое подразделение суда

/update_status [номер дела] [статус]: изменение статуса искового заявления

/search_claim [критерии поиска]: поиск искового заявления по различным критериям


### Пример 2

Команды для обработки искового заявления
1. Прием и регистрация

ПРИНЯТЬ — зафиксировать поступление искового заявления.
ЗАРЕГИСТРИРОВАТЬ — внести заявление в базу данных.

2. Проверка заявления

ПРОВЕРИТЬ ПОЛНОТУ — убедиться в наличии всех документов и данных.
ПРОВЕРИТЬ ПРАВИЛЬНОСТЬ — проверить корректность заполнения.

3. Определение подсудности

ОПРЕДЕЛИТЬ ПОДСУДНОСТЬ — установить суд, который будет рассматривать дело.
ОПРЕДЕЛИТЬ ПОДВЕДОМСТВЕННОСТЬ — проверить компетенцию суда для рассмотрения дела.

4. Определение порядка производства

ОПРЕДЕЛИТЬ ПОРЯДОК — выбрать вид производства: исковое, особое, приказное.

5. Уведомление сторон

УВЕДОМИТЬ ИСТЦА — проинформировать истца о принятии заявления.
УВЕДОМИТЬ ОТВЕТЧИКА — уведомить ответчика о начале процесса и запросить отзыв.

6. Назначение заседания

НАЗНАЧИТЬ ЗАСЕДАНИЕ — определить дату и время судебного разбирательства.

7. Дополнительные действия

ИСТРЕБОВАТЬ ДОКУМЕНТЫ — запросить недостающие материалы.
НАПРАВИТЬ ЗАПРОСЫ — отправить запросы в другие организации.
ПРИНЯТЬ МЕРЫ — обеспечить исполнение иска при необходимости.

Алгоритмическая система – набор средств и понятий, позволяющих строить некоторое множество алгоритмов для решения определенного класса задач. Алгоритмическая система определяется наличием четырех составляющих ее частей:

1) множеством входных объектов или исходных данных, подлежащих обработке алгоритмами данной системы;

2) множеством выходных объектов или результатов выполнения алгоритмов данной системы;

3) системой команд исполнителя, то есть набором тех действий, которые может выполнять исполнитель, и которые мы можем описывать в алгоритмах, что собственно является ориентацией алгоритмической системы на конкретного исполнителя;

4) языком описания алгоритмов – языком исполнителя; язык, на котором описан алгоритм, должен быть понятен исполнителю и не должен включать в свой состав указания на невозможные для исполнителя действия, а также обращения к входным или выходным объектам, не принадлежащих к множеству входных или выходных объектов данной алгоритмической системы.
