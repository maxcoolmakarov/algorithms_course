## Типы данных

````python
import sys

# Примеры различных типов данных
data_examples = {
    "Integer": 42,
    "Float": 3.14,
    "String": "Hello, World!",
    "List": [1, 2, 3, 4, 5],
    "Dictionary": {"a": 1, "b": 2, "c": 3}
}

# Выводим размер и адрес в памяти для каждого типа данных
for data_type, value in data_examples.items():
    memory_size = sys.getsizeof(value)  # Размер в байтах
    memory_address = id(value)  # Адрес в памяти
    print(f"{data_type} - Значение: {value}")
    print(f"  Размер: {memory_size} байт")
    print(f"  Адрес: {memory_address}\n")
````

В Python элементы списка не всегда хранятся в соседних ячейках памяти, так как Python списки (список — это объект типа list) представляют собой массив указателей на объекты. Это значит, что Python хранит ссылки на элементы списка, а не сами значения в последовательных ячейках. Эти ссылки могут указывать на объекты, которые физически расположены в разных участках памяти.

Зато в C++ все по честному
```C
#include <iostream>

int main() {
    int arr[5] = {10, 20, 30, 40, 50};  // Создаем массив из 5 целых чисел

    // Выводим адрес каждого элемента массива
    for (int i = 0; i < 5; i++) {
        std::cout << "Элемент " << i << ": Значение = " << arr[i] 
                  << ", Адрес = " << &arr[i] << std::endl;
    }

    int* ptr = arr;  // Указатель на первый элемент массива
    for (int i = 0; i < 5; i++) {
        std::cout << "Адрес ptr + " << i << " = " << (ptr + i) << " элемент " << *(ptr + i) << std::endl;
    }


    return 0;
}
```


## Рекурсия 
Метод программирования, в котором функция вызывает сама себя для решения подзадач. Рекурсия особенно полезна в задачах, которые можно разбить на похожие подзадачи.

При использовании рекурсии важно иметь условие выхода — условие, при котором рекурсивный вызов завершится. Без него рекурсия будет бесконечной и приведет к ошибке переполнения стека.

1. Факториал числа

    Факториал числа n (обозначается как n!) — это произведение всех чисел от 1 до n.
    
    Рекурсивное определение:
    n!=n×(n−1)!
    
    где 1!=1.

````python
6! = 6 * (5 * 4 * 3 * 2 * 1)
````

````python
f(5) = 5 * f(4)
f(1) = 1
````

````python
def factorial(n):
    if n == 1:  # Условие выхода
        return 1
    return n * factorial(n - 1)  # Рекурсивный вызов
````

````python
factorial(4)
 4 * factorial(3)

     factorial(3)
     3 * factorial(2)
         factorial(2)

          2 * factorial(1)
````

2. Числа Фибоначчи

    Последовательность Фибоначчи определяется как:
    F(0)=0,F(1)=1
    
    и для всех n>1:
    F(n)=F(n−1)+F(n−2)

````python
def fibonacci(n):
    if n == 0:  # Условие выхода
        return 0
    elif n == 1:  # Условие выхода
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)  # Рекурсивный вызов
````

## Система команд

Совокупность инструкций, которые процессор может выполнять. Эти команды включают базовые операции, такие как арифметические и логические действия, управление памятью, ввод/вывод, а также управление потоком выполнения программы. Каждая команда состоит из кода операции и операндов (данных или адресов), к которым применяется операция. Процессор интерпретирует и выполняет команды, используя свою архитектуру и схему взаимодействия с памятью и периферийными устройствами.

Основная цель системы команд — предоставить разработчикам и процессорам интерфейс для выполнения задач на аппаратном уровне. Система команд определяет, какие операции можно выполнять, и, таким образом, ограничивает возможности процессора в выполнении определенных задач. Она играет роль посредника между программным обеспечением и аппаратным обеспечением, обеспечивая выполнение инструкций, написанных в машинном коде или ассемблере.

### Примеры систем команд

#### CISC (Complex Instruction Set Computer) — Компьютер со сложной системой команд

CISC-системы команд включают широкий набор сложных инструкций, каждая из которых может выполнять несколько шагов. Это позволяет сократить количество инструкций, требуемых для выполнения сложных операций, но такие команды часто сложны и требуют большего времени на выполнение.
* Примеры процессоров: x86 (Intel и AMD).
* Примеры применения: Используется в персональных компьютерах, где важна совместимость с устаревшими системами и высокая производительность на сложных операциях.

#### RISC (Reduced Instruction Set Computer) — Компьютер с сокращенной системой команд

Описание: В RISC архитектурах используется минимальный набор простых команд. Каждая команда выполняется за один цикл, что упрощает процессор и делает его быстрее и эффективнее в плане энергопотребления.
* Примеры процессоров: ARM (в мобильных устройствах), MIPS (раньше использовался в суперкомпьютерах и встраиваемых системах), PowerPC.
* Примеры применения: Мобильные устройства, встраиваемые системы, где важна высокая производительность при малом потреблении энергии.

#### GPU (Graphics Processing Unit) — Графический процессор

Описание: Система команд GPU предназначена для массового параллельного выполнения операций. Эти процессоры ориентированы на высокоэффективное выполнение операций с матрицами и векторами, необходимых для графической обработки и машинного обучения.
* Примеры процессоров: NVIDIA CUDA, AMD Radeon.
* Примеры применения: Компьютерная графика, машинное обучение, обработка больших данных.
