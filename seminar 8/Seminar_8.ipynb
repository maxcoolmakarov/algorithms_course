## Алгоритм Хаффмана
    Жадный алгоритм сжатия данных без потерь, используемый для кодирования символов таким образом, чтобы часто встречающиеся символы занимали меньше места, а редкие — больше. Он особенно полезен, когда нужно эффективно сохранить или передать данные (например, в форматах PNG, MP3, ZIP).

### Зачем появился алгоритм Хаффмана?

До Хаффмана данные часто кодировали фиксированной длиной — например, 1 байт (8 бит) на символ. Но это неэффективно, если в тексте часто повторяются одни и те же символы.

Цель алгоритма Хаффмана — уменьшить общий размер закодированных данных, используя переменную длину кодов:

    Частые символы → короткие коды

    Редкие символы → длинные коды

### Как работает алгоритм Хаффмана?
1. Подсчитать частоты всех символов.
   
    ```
    A: 15 раз  
    B: 7 раза  
    C: 6 раз  
    D: 6 раз
    E: 5 раз
    ```
2. Создать приоритетную очередь (или список), где каждый элемент — узел с символом и его частотой.

   
4. Построить дерево Хаффмана:

    Пока в очереди больше одного узла:

   1. Взять два узла с наименьшей частотой
   3. Объединить их в новый узел (частота = сумма двух)
   4. Поместить новый узел обратно в очередь

      

       Теперь наша очередь будет 15, 11, 7, 6 и нужно взять узлы 7 и 6:

      

       Теперь наша очередь будет 15, 13, 11 и нужно взять узлы 13 и 11 (которые являются вершинами для других ранее посчитанных):
      
      
      
      
4. Назначить коды:

    - Обход дерева от корня:
  
        Левый путь → 0
  
        Правый путь → 1

    - Каждому листу (символу) присваивается путь от корня до него
  
    

5. Закодировать строку:
   
   Каждую букву заменить её Хаффмановским кодом.
